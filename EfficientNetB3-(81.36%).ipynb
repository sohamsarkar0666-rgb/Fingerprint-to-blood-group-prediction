```python
"""
=============================================================
  Fingerprint → Blood Group  |  EfficientNetB3

  Why EfficientNetB3 for fingerprints:
    - Better than V2S for small datasets (original EfficientNet
      was designed for smaller data regimes)
    - 81.6% ImageNet top-1 — strong feature extractor
    - 12.3M params — compact, fast on T4
    - Compound scaling balances width/depth/resolution
      without over-engineering like V2 variants
=============================================================
"""

# Colab setup (uncomment if needed):
# !pip install opendatasets -q
# import opendatasets as od
# od.download('https://www.kaggle.com/datasets/rohitpravinlohar/fingerprint-dataset-for-blood-group-classification')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.applications import EfficientNetB3
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras import callbacks
from sklearn.metrics import classification_report, confusion_matrix

# ─────────────────────────────────────────────
#  CONFIG
# ─────────────────────────────────────────────
DATA_DIR    = r'/kaggle/working/fingerprint-to-blood-group-prediction/main_dataset_after_split/Fingerprint Dataset for Blood Group Classification'
IMG_SIZE    = (256, 256)
BATCH_SIZE  = 32
EPOCHS      = 60
NUM_CLASSES = 8
SAVE_PATH   = 'best_efficientnetb3.keras'
MODEL_NAME  = 'EfficientNetB3'

tf.random.set_seed(42)
np.random.seed(42)

# ─────────────────────────────────────────────
#  DATA
#  EfficientNet preprocess_input scales [0,255] → [0,1]
#  with its own normalization (different from V2)
# ─────────────────────────────────────────────
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.20,
    horizontal_flip=True,
    rotation_range=10,
    zoom_range=0.05,
)
val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.20,
)

train_gen = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True,
    seed=42,
)
val_gen = val_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False,
    seed=42,
)

class_names = list(train_gen.class_indices.keys())
print(f"Train: {train_gen.samples} | Val: {val_gen.samples}")
print(f"Classes: {class_names}")

# ─────────────────────────────────────────────
#  MODEL
# ─────────────────────────────────────────────
base = EfficientNetB3(
    input_shape=(256, 256, 3),
    include_top=False,
    weights='imagenet',
    pooling='avg'
)

# Freeze all first
for layer in base.layers:
    layer.trainable = False

# Unfreeze top 20 layers
# EfficientNetB3 is more compact than V2S so
# top 20 covers the last MBConv block well
for layer in base.layers[-20:]:
    layer.trainable = True

# Head
x      = Dense(256, activation='relu')(base.output)
x      = BatchNormalization()(x)
x      = Dropout(0.3)(x)
x      = Dense(128, activation='relu')(x)
x      = Dropout(0.2)(x)
output = Dense(NUM_CLASSES, activation='softmax')(x)

model = Model(inputs=base.input, outputs=output)

# ─────────────────────────────────────────────
#  COMPILE
# ─────────────────────────────────────────────
total_steps = (train_gen.samples // BATCH_SIZE) * EPOCHS
lr_schedule = tf.keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=1e-3,
    decay_steps=total_steps,
    alpha=1e-6,
)

model.compile(
    optimizer=AdamW(learning_rate=lr_schedule, weight_decay=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

trainable = sum(tf.size(w).numpy() for w in model.trainable_weights)
total     = sum(tf.size(w).numpy() for w in model.weights)
print(f"\n{MODEL_NAME} — Trainable: {trainable:,} | Total: {total:,}\n")

# ─────────────────────────────────────────────
#  CALLBACKS
# ─────────────────────────────────────────────
cb_list = [
    callbacks.ModelCheckpoint(
        SAVE_PATH, monitor='val_accuracy',
        save_best_only=True, verbose=1,
    ),
    callbacks.EarlyStopping(
        monitor='val_accuracy', patience=15,
        restore_best_weights=True, verbose=1,
    ),

]

# ─────────────────────────────────────────────
#  TRAIN
# ─────────────────────────────────────────────
history = model.fit(
    train_gen,
    epochs=EPOCHS,
    validation_data=val_gen,
    callbacks=cb_list,
    verbose=1,
)

best_val = max(history.history['val_accuracy']) * 100
print(f"\nBest val_accuracy: {best_val:.2f}%")

# ─────────────────────────────────────────────
#  EVALUATE
# ─────────────────────────────────────────────
val_gen.reset()
y_pred = np.argmax(model.predict(val_gen, verbose=1), axis=1)
y_true = val_gen.classes

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# ─────────────────────────────────────────────
#  PLOTS
# ─────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(20, 5))
fig.patch.set_facecolor('#0d1117')
C = {"tr":"#58a6ff","va":"#f97583","tx":"#c9d1d9","gr":"#21262d","bg":"#161b22","lr":"#3fb950"}

def style(ax, title, xlabel, ylabel):
    ax.set_facecolor(C["bg"])
    ax.set_title(title, color=C["tx"], fontsize=13, fontweight="bold", pad=8)
    ax.set_xlabel(xlabel, color=C["tx"])
    ax.set_ylabel(ylabel, color=C["tx"])
    ax.tick_params(colors=C["tx"])
    for sp in ax.spines.values(): sp.set_edgecolor(C["gr"])
    ax.grid(color=C["gr"], lw=0.6, ls="--", alpha=0.7)
    ax.legend(fontsize=9, facecolor=C["bg"], labelcolor=C["tx"])

ep = range(1, len(history.history['loss']) + 1)
axes[0].plot(ep, history.history['loss'],     color=C["tr"], lw=2, label='Train')
axes[0].plot(ep, history.history['val_loss'], color=C["va"], lw=2, label='Val')
style(axes[0], "Loss Curve", "Epoch", "Loss")

axes[1].plot(ep, [a*100 for a in history.history['accuracy']],     color=C["tr"], lw=2, label='Train')
axes[1].plot(ep, [a*100 for a in history.history['val_accuracy']], color=C["va"], lw=2, label='Val')
axes[1].axhline(y=best_val, color=C["lr"], ls='--', lw=1.2, label=f'Best: {best_val:.1f}%')
style(axes[1], "Accuracy Curve", "Epoch", "Accuracy (%)")

cm_mat = confusion_matrix(y_true, y_pred)
cm_n   = cm_mat.astype("float") / cm_mat.sum(axis=1, keepdims=True)
sns.heatmap(cm_n, annot=cm_mat, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names,
            ax=axes[2], annot_kws={"size": 9}, linewidths=0.5)
axes[2].set_title("Confusion Matrix", color=C["tx"], fontsize=13, fontweight="bold", pad=8)
axes[2].set_xlabel("Predicted", color=C["tx"])
axes[2].set_ylabel("True", color=C["tx"])
axes[2].tick_params(colors=C["tx"])
axes[2].set_facecolor(C["bg"])

fig.suptitle(f"{MODEL_NAME} | AdamW + Cosine Decay | Best Val: {best_val:.2f}%",
             color=C["tx"], fontsize=12, fontweight="bold")
plt.tight_layout()
plt.savefig(f'training_report_{MODEL_NAME.lower()}.png', dpi=150,
            bbox_inches='tight', facecolor=fig.get_facecolor())
plt.show()
print(f"Saved → training_report_{MODEL_NAME.lower()}.png")

# from google.colab import files
# files.download(SAVE_PATH)
```

```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.efficientnet import preprocess_input
from sklearn.metrics import classification_report, confusion_matrix

# ─────────────────────────────────────────────
#  LOAD MODEL
# ─────────────────────────────────────────────
model = load_model('/kaggle/working/best_efficientnetb3.keras')

# ─────────────────────────────────────────────
#  PREDICTION DATASET
# ─────────────────────────────────────────────
pred_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

pred_gen = pred_datagen.flow_from_directory(
    directory="/kaggle/working/fingerprint-to-blood-group-prediction/prediction_dataset",
    target_size=(256, 256),  # ✅ same as training
    batch_size=32,
    class_mode='categorical',
    shuffle=False,            # important for correct label order
)

# ─────────────────────────────────────────────
#  MAKE PREDICTIONS
# ─────────────────────────────────────────────
y_pred_probs = model.predict(pred_gen, verbose=1)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = pred_gen.classes
class_names = list(pred_gen.class_indices.keys())

# ─────────────────────────────────────────────
#  REPORT
# ─────────────────────────────────────────────
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

cm = confusion_matrix(y_true, y_pred)
print("\nConfusion Matrix:\n", cm)
```

```python
# ─────────────────────────────────────────────
#  CONFUSION MATRIX HEATMAP
# ─────────────────────────────────────────────
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_true, y_pred)

# Normalize (optional but useful)
cm_norm = cm.astype("float") / cm.sum(axis=1, keepdims=True)

plt.figure(figsize=(8, 6))

sns.heatmap(
    cm_norm,
    annot=cm,              # show actual numbers
    fmt="d",
    cmap="Blues",
    xticklabels=class_names,
    yticklabels=class_names,
    linewidths=0.5,
    annot_kws={"size": 10}
)

plt.title("Confusion Matrix (Prediction Dataset)", fontsize=14, fontweight="bold")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```
